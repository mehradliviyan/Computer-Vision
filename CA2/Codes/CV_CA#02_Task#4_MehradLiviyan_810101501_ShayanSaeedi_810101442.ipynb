## Correlation


<div dir="rtl">

در این روش، ایده اصلی این است که بررسی کنیم تغییرات شدت پیکسل‌های دو تصویر چقدر با هم هم‌جهت هستند.
به زبان ساده‌تر، اگر در هر دو تصویر پیکسل‌ها در جاهای مشابه روشن‌تر یا تیره‌تر شوند، مقدار همبستگی (Correlation) بالا خواهد بود و این یعنی تصاویر از نظر ساختار و بافت به هم شبیه‌اند. اما اگر در یک تصویر نواحی روشن در جایی باشند که در تصویر دیگر تیره‌اند، مقدار همبستگی پایین می‌آید یا حتی منفی می‌شود.

فرمول کلی این روش به شکل زیر است:

</div>

<div dir="ltr">

$$
r = \frac{\sum (A_i - \bar{A})(B_i - \bar{B})}{\sqrt{\sum (A_i - \bar{A})^2 \sum (B_i - \bar{B})^2}}
$$

</div>

<div dir="rtl">

در این رابطه، $(A_i)$ و $(B_i)$ شدت پیکسل‌های متناظر در دو تصویر هستند، و $(\bar{A})$ و $(\bar{B})$ میانگین شدت پیکسل‌ها در هر تصویر را نشان می‌دهند. عبارت ((A_i - \bar{A})) در واقع اختلاف هر پیکسل از میانگین روشنایی تصویر است، که باعث می‌شود تأثیر روشن‌تر یا تیره‌تر بودن کلی تصویر حذف شود و فقط "الگوی تغییرات" مورد مقایسه قرار گیرد.

مقدار خروجی (r) بین -1 تا 1 است:
عدد 1 نشان‌دهنده‌ی شباهت کامل، عدد 0 نشان‌دهنده‌ی عدم ارتباط، و عدد -1 به معنی برعکس بودن کامل دو تصویر است.

در مجموع، این روش بسیار ساده و سریع است، اما نسبت به تغییر روشنایی یا کنتراست بین دو تصویر حساس است؛ یعنی اگر یکی از تصاویر کمی روشن‌تر یا تاریک‌تر باشد، مقدار شباهت ممکن است کاهش پیدا کند، حتی اگر ساختار کلی دو تصویر یکسان باشد.


````python
import numpy as np
import pandas as pd

CSV_PATH = "../Results/Segment_Correlation.csv"

results = []

for img_id, char_list in segmented_chars.items():
    for char_idx, char in enumerate(char_list):
        char_img = char['image']
        best_corr = -1
        predicted_label = None

        for ref in mapset_images:
            ref_img = ref['image']

            if char_img.shape != ref_img.shape:
                char_resized = cv2.resize(char_img, (ref_img.shape[1], ref_img.shape[0]), interpolation=cv2.INTER_NEAREST)
                print("Unsized")
            else:
                char_resized = char_img

            corr = np.corrcoef(char_resized.flatten(), ref_img.flatten())[0,1]

            if corr > best_corr:
                best_corr = corr
                predicted_label = ref['name']

        output_name = f"Image_{img_id}_Char_{char_idx+1:02d}.png"

        results.append({
            "File": output_name,
            "Correlation": best_corr,
            "Predicted_Label": predicted_label
        })

df = pd.DataFrame(results)
df.to_csv(CSV_PATH, index=False)
print(f"CSV saved at {CSV_PATH}")
````

## MSE

<div dir="rtl">

در این روش، ایده اصلی این است که میزان اختلاف بین دو تصویر را با محاسبه‌ی میانگین مربعات تفاضل پیکسل‌ها اندازه بگیریم.
به زبان ساده‌تر، اگر دو تصویر دقیقاً شبیه هم باشند، اختلاف شدت پیکسل‌های آن‌ها صفر است و در نتیجه مقدار MSE هم صفر می‌شود.
اما هر چه مقدار MSE بزرگ‌تر باشد، یعنی تفاوت بین تصاویر بیشتر است و شباهت آن‌ها کمتر خواهد بود.

فرمول کلی این روش به شکل زیر است:

</div>

<div dir="ltr">

$$
MSE = \frac{1}{N} \sum_{i=1}^{N} (A_i - B_i)^2
$$

</div>

<div dir="rtl">

در این رابطه، (A_i) و (B_i) شدت پیکسل‌های متناظر در دو تصویر هستند و (N) تعداد کل پیکسل‌هاست.
در واقع ما برای هر پیکسل، اختلاف مقدارش را نسبت به پیکسل مشابه در تصویر دوم می‌گیریم، آن را به توان دو می‌رسانیم (تا اختلاف‌های مثبت و منفی با هم جمع نشوند) و در نهایت میانگین تمام این مقدارها را محاسبه می‌کنیم.

در این روش، هر چه مقدار MSE به صفر نزدیک‌تر باشد، یعنی تصاویر شباهت بیشتری دارند.
برعکس، مقدار MSE بالا نشان‌دهنده تفاوت زیاد بین دو تصویر است.

به طور کلی، MSE یک روش بسیار ساده، سریع و قابل فهم است، اما یک ایراد بزرگ دارد:
به ساختار یا درک بصری انسان توجهی ندارد. یعنی ممکن است دو تصویر از نظر MSE تفاوت زیادی داشته باشند، در حالی که از دید انسان تقریباً یکسان به نظر برسند (مثلاً وقتی یکی از آن‌ها کمی نویز دارد).

به همین خاطر معمولاً از MSE در کنار معیارهای دقیق‌تر مثل SSIM استفاده می‌شود تا ارزیابی واقعی‌تری از شباهت تصاویر به دست آید.


````python
CSV_PATH = "../Results/Segment_MSE.csv"

results = []

def compute_mse(img1, img2):
    if img1.shape != img2.shape:
        img1 = cv2.resize(img1, (img2.shape[1], img2.shape[0]), interpolation=cv2.INTER_NEAREST)
    return np.mean((img1.astype("float") - img2.astype("float")) ** 2)

for img_id, char_list in segmented_chars.items():
    for char_idx, char in enumerate(char_list):
        char_img = char["image"]
        best_mse = float("inf")
        predicted_label = None

        for ref in mapset_images:
            ref_img = ref["image"]

            mse_value = compute_mse(char_img, ref_img)

            if mse_value < best_mse:
                best_mse = mse_value
                predicted_label = ref["name"]

        output_name = f"Image_{img_id}_Char_{char_idx+1:02d}.png"

        results.append({
            "File": output_name,
            "MSE": best_mse,
            "Predicted_Label": predicted_label
        })

df = pd.DataFrame(results)
df.to_csv(CSV_PATH, index=False)
print(f"CSV saved at {CSV_PATH}")
````

## SSIM

<div dir="rtl">


ایده‌ی اصلی این روش اینه که شباهت بین دو تصویر فقط با اختلاف پیکسل‌ها مشخص نمی‌شه،
بلکه باید ساختار، روشنایی و کنتراست تصویر هم در نظر گرفته بشه.
به زبان ساده‌تر، **SSIM** سعی می‌کنه شباهت دو تصویر رو از دید انسان بررسی کنه، نه فقط از نظر عددی.

این روش سه جنبه‌ی اصلی تصویر رو مقایسه می‌کنه:

1. روشنایی (Luminance): میانگین شدت پیکسل‌ها در دو تصویر
2. کنتراست (Contrast): تغییرات یا انحراف معیار شدت پیکسل‌ها
3. ساختار (Structure): نحوه‌ی توزیع و الگوی پیکسل‌ها در تصویر

فرمول کلی SSIM به شکل زیر است:

</div>

<div dir="ltr">

$$
SSIM(A, B) = \frac{(2\mu_A \mu_B + C_1)(2\sigma_{AB} + C_2)}{(\mu_A^2 + \mu_B^2 + C_1)(\sigma_A^2 + \sigma_B^2 + C_2)}
$$

</div>

<div dir="rtl">

در این رابطه:

* $( \mu_A )$ و $( \mu_B )$: میانگین شدت پیکسل‌های هر تصویر
* $( \sigma_A^2 )$ و $( \sigma_B^2 )$: واریانس یا میزان تغییرات شدت در هر تصویر
* $( \sigma_{AB} )$: کوواریانس بین دو تصویر (نشان‌دهنده‌ی میزان هم‌جهتی تغییرات)
* $( C_1 )$ و $( C_2 )$: مقادیر ثابت کوچکی برای جلوگیری از تقسیم بر صفر

مقدار SSIM همیشه بین 0 و 1 است:

* **1** یعنی دو تصویر کاملاً شبیه هم‌اند
* **0** یعنی هیچ شباهتی ندارند

به طور کلی، SSIM از نظر ادراکی خیلی دقیق‌تر از روش‌های ساده‌ای مثل MSE یا Correlation عمل می‌کند،
چون سعی می‌کند شباهت بین تصاویر را همان‌طور که چشم انسان می‌بیند، بسنجد.
به همین دلیل در ارزیابی کیفیت تصویر، فشرده‌سازی، یا بازسازی تصویر، **SSIM یکی از دقیق‌ترین معیارها** به حساب می‌آید.


````python
import cv2
import numpy as np
import pandas as pd
from skimage.metrics import structural_similarity as ssim

CSV_PATH = "../Results/Segment_SSIM.csv"

results = []

def compute_ssim(img1, img2):
    if len(img1.shape) == 3:
        img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
    if len(img2.shape) == 3:
        img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

    if img1.shape != img2.shape:
        img1 = cv2.resize(img1, (img2.shape[1], img2.shape[0]), interpolation=cv2.INTER_NEAREST)

    score, _ = ssim(img1, img2, full=True)
    return score

for img_id, char_list in segmented_chars.items():
    for char_idx, char in enumerate(char_list):
        char_img = char["image"]
        best_ssim = -1
        predicted_label = None

        for ref in mapset_images:
            ref_img = ref["image"]
            ssim_value = compute_ssim(char_img, ref_img)

            if ssim_value > best_ssim:
                best_ssim = ssim_value
                predicted_label = ref["name"]

        output_name = f"Image_{img_id}_Char_{char_idx+1:02d}.png"

        results.append({
            "File": output_name,
            "SSIM": best_ssim,
            "Predicted_Label": predicted_label
        })

df = pd.DataFrame(results)
df.to_csv(CSV_PATH, index=False)
print(f"CSV saved at {CSV_PATH}")
````

## Histogram Compare

<div dir="rtl">

ایده‌ی اصلی این روش اینه که به‌جای بررسی پیکسل‌به‌پیکسل دو تصویر، بیایم **توزیع رنگ یا شدت روشنایی** پیکسل‌ها رو با هم مقایسه کنیم.
به زبان ساده‌تر، به‌جای اینکه ببینیم هر پیکسل دقیقاً چقدره، بررسی می‌کنیم **چند تا پیکسل روشن، متوسط یا تیره** در هر تصویر وجود داره و این الگوها رو با هم مقایسه می‌کنیم.

در واقع، هر تصویر رو به شکل یه نمودار (هیستوگرام) در نظر می‌گیریم که محور افقی شدت روشنایی (۰ تا ۲۵۵) و محور عمودی تعداد پیکسل‌های با اون شدت رو نشون می‌ده.
اگر دو تصویر هیستوگرام‌های مشابهی داشته باشن، یعنی از نظر توزیع روشنایی و کنتراست شبیه هم‌ان—even اگر دقیقاً پیکسل‌به‌پیکسل یکی نباشن.

فرمول کلی شباهت دو هیستوگرام می‌تونه به شکل‌های مختلفی محاسبه بشه، یکی از رایج‌ترین‌هاش **همبستگی هیستوگرام‌ها** است:


</div>

<div dir="ltr">

$$
r = \frac{\sum (H_1(i) - \bar{H_1})(H_2(i) - \bar{H_2})}{\sqrt{\sum (H_1(i) - \bar{H_1})^2 \sum (H_2(i) - \bar{H_2})^2}}
$$

</div>

<div dir="rtl">

که در آن:

* $( H_1(i) )$ و $( H_2(i) )$: مقدار هیستوگرام در سطح شدت ( i ) برای دو تصویر
* $( \bar{H_1} )$ و $( \bar{H_2} )$: میانگین مقادیر هیستوگرام‌ها

بازه مقدار معمولاً بین -1 تا 1 است:

* **1** یعنی دو هیستوگرام (و در نتیجه دو تصویر) خیلی شبیه‌اند
* **0** یعنی هیچ شباهتی ندارند
* **-1** یعنی برعکس هم‌اند

این روش به‌خصوص وقتی خوب جواب می‌ده که هدف، مقایسه‌ی **توزیع کلی روشنایی یا رنگ تصویر** باشه (مثلاً در تطبیق صحنه‌ها یا تشخیص شباهت کلی بین تصاویر).
اما از اونجایی که مکان پیکسل‌ها در این روش در نظر گرفته نمی‌شه، دو تصویر ممکنه هیستوگرام مشابهی داشته باشن ولی از نظر ساختار کاملاً متفاوت باشن.


````python
CSV_PATH = "../Results/Segment_Histogram.csv"

results = []

def histogram_similarity(img1, img2):
    if len(img1.shape) == 3:
        img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
    if len(img2.shape) == 3:
        img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
    if img1.shape != img2.shape:
        img1 = cv2.resize(img1, (img2.shape[1], img2.shape[0]), interpolation=cv2.INTER_NEAREST)
    hist1 = cv2.calcHist([img1], [0], None, [256], [0, 256])
    hist2 = cv2.calcHist([img2], [0], None, [256], [0, 256])
    cv2.normalize(hist1, hist1)
    cv2.normalize(hist2, hist2)
    similarity = cv2.compareHist(hist1, hist2, cv2.HISTCMP_CORREL)
    return similarity

for img_id, char_list in segmented_chars.items():
    for char_idx, char in enumerate(char_list):
        char_img = char["image"]
        best_sim = -1
        predicted_label = None

        for ref in mapset_images:
            ref_img = ref["image"]
            sim_value = histogram_similarity(char_img, ref_img)

            if sim_value > best_sim:
                best_sim = sim_value
                predicted_label = ref["name"]

        output_name = f"Image_{img_id}_Char_{char_idx+1:02d}.png"

        results.append({
            "File": output_name,
            "Histogram_Correlation": best_sim,
            "Predicted_Label": predicted_label
        })

df = pd.DataFrame(results)
df.to_csv(CSV_PATH, index=False)
print(f"CSV saved at {CSV_PATH}")
````

## Template Matching

<div dir="rtl">


ایده‌ی اصلی این روش اینه که یه تصویر (یا بخشی از تصویر) رو به‌عنوان **الگو (Template)** در نظر بگیریم و اون رو با تصویر دوم مقایسه کنیم تا ببینیم چقدر باهاش مطابقت داره.
در حالت کلی، این روش برای پیدا کردن یه الگو در داخل یه تصویر بزرگ‌تر استفاده می‌شه (مثل پیدا کردن یه حرف یا نماد خاص در صفحه)، ولی وقتی اندازه‌ی دو تصویر یکی باشه، عملاً کارش مثل **محاسبه‌ی همبستگی بین دو تصویر هم‌اندازه** است.

در این حالت (که تصاویر ما همه 64×64 هستن)، Template Matching با روش Correlation از نظر محاسباتی خیلی شبیهه، چون مستقیماً پیکسل‌های متناظر دو تصویر رو با هم مقایسه می‌کنه تا میزان شباهت اون‌ها رو بسنجیم.

فرمول کلی این روش در حالت نرمال‌سازی‌شده به شکل زیره:

</div>

<div dir="ltr">

$$
R(x, y) = \frac{\sum_{i,j} [T(i,j) - \bar{T}][I(x+i, y+j) - \bar{I}*{x,y}]}{\sqrt{\sum*{i,j}[T(i,j) - \bar{T}]^2 \sum_{i,j}[I(x+i, y+j) - \bar{I}_{x,y}]^2}}
$$


</div>

<div dir="rtl">

که در اون:

* $( T(i,j) )$: شدت پیکسل در تصویر الگو (Template)
* $( I(x+i, y+j) )$: شدت پیکسل در تصویر مورد بررسی
* $( \bar{T} )$ و $( \bar{I}_{x,y} )$: میانگین شدت پیکسل‌ها در هر ناحیه

اما چون در پروژه‌ی ما هر دو تصویر دقیقاً هم‌اندازه هستن، عبارت بالا فقط یه مقدار شباهت عددی بین دو تصویر برمی‌گردونه و نیازی به جابه‌جایی (Sliding) نیست.

خروجی این روش معمولاً بین **-1 تا 1** نرمال می‌شه:

* **1** → دو تصویر دقیقاً یکسان
* **0** → هیچ شباهتی ندارن
* **-1** → کاملاً برعکس هم‌ان

در نتیجه، در این پروژه روش Template Matching از نظر عملکرد خیلی نزدیک به Correlationه، فقط نحوه‌ی پیاده‌سازی در OpenCV (تابع `cv2.matchTemplate`) فرق داره.
اما چون اندازه‌ی تصاویر یکیه، هر دو روش تقریباً **خروجی مشابهی** خواهند داد.


````python
import cv2
import numpy as np
import pandas as pd

CSV_PATH = "../Results/Segment_TemplateMatching.csv"

results = []

for img_id, char_list in segmented_chars.items():
    for char_idx, char in enumerate(char_list):
        char_img = char['image']
        best_score = -1
        predicted_label = None

        for ref in mapset_images:
            ref_img = ref['image']

            if char_img.shape[0] > ref_img.shape[0] or char_img.shape[1] > ref_img.shape[1]:
                char_resized = cv2.resize(char_img, (ref_img.shape[1], ref_img.shape[0]), interpolation=cv2.INTER_NEAREST)
                template = char_resized
                image = ref_img
            else:
                template = char_img
                image = ref_img
                
            res = cv2.matchTemplate(image, template, cv2.TM_CCOEFF_NORMED)
            _, max_val, _, _ = cv2.minMaxLoc(res)

            if max_val > best_score:
                best_score = max_val
                predicted_label = ref['name']

        output_name = f"Image_{img_id}_Char_{char_idx+1:02d}.png"

        results.append({
            "File": output_name,
            "Template_Score": best_score,
            "Predicted_Label": predicted_label
        })

df = pd.DataFrame(results)
df.to_csv(CSV_PATH, index=False)
print(f"CSV saved at {CSV_PATH}")
````
